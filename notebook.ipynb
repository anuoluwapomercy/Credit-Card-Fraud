```python
# To analyse whether the older customers are likely to be victims of credit card 
import pandas as pd
from datetime import datetime

# Load the dataset
ccf = pd.read_csv('credit_card_fraud.csv')

# Calculate age
ccf['age'] = (datetime.now() - pd.to_datetime(ccf['dob'])).dt.days / 365.25

# Create age groups
ccf['age_group'] = pd.cut(ccf['age'], bins=[0, 30, 60, float('inf')], 
                          labels=['Young', 'Middle-aged', 'Senior'])

# Calculate fraud rate per age group
fraud_rate_by_age = ccf.groupby('age_group')['is_fraud'].mean()

# Print the results
print(fraud_rate_by_age)
```
